```python
import re
import pandas as pd
```

```python
f = open("chat3_12hr.txt",'r',encoding='utf-8')
```

```python
data = f.read()
data
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
msg = re.split(pattern,data)[1:]
msg
```

```python
len(msg)
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_msg' : msg, 'msg_date' : dates})

## Converting datatype of dates

df['msg_date'] = pd.to_datetime(df['msg_date'], format='%m/%d/%y, %H:%M - ')

df.rename(columns={'msg_date':'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
## Separating Users and messages

users = []
messages = []

for msg in df['user_msg']:
    entry = re.split('([\w\W]+?):\s',msg)
    if entry[1:]:# username
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('grp_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.head()
        
```
